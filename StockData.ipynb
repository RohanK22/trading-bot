```
from pandas_datareader import data as web
import datetime
```

```
stock = 'SPY'
```

```
start_date = (datetime.datetime.now() - datetime.timedelta(days=365)).strftime("%m-%d-%Y")
df = web.DataReader(stock, data_source='yahoo', start=start_date)
```
